```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import pandas_datareader as web
from pandas.plotting import lag_plot
from datetime import datetime
from sklearn.ensemble import RandomForestRegressor
#from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
from datetime import datetime
import yfinance as yf
from sklearn.model_selection import train_test_split
from sklearn import metrics
import pandas_datareader.data as pdr
import statsmodels.api as smapi
```

```python
df = pd.read_csv('C:/Users/abhi/Desktop/ARIMA_All/ARIMA_All/Train_Status.csv')
df.columns
```

```python
out = []
pred=[]
```

```python
def smape_kun(y_true, y_pred):
    return np.mean((np.abs(y_pred - y_true) * 200/ (np.abs(y_pred) + np.abs(y_true))))
```

```python
# select column "Close" as the sample 
for i in df["Name of Stock"]:
        print("Running Random Forest for :"+i)
        try:
#             print("In try")
            data = pd.read_csv("C:/Users/abhi/Desktop/ARIMA_All/ARIMA_All/individual_stocks_5yr/"+i+"_data.csv")
#             data = data.dropna(how='any',axis=0)
#             X  = data[['open','high','low','volume']]
#             y = data['close']
#             print(y)
            #df2 = df.iloc[: , [1, 2]].copy()
#             data1 = data['close'].values
        except:
#             print("In except")
            start = datetime.strptime('2014-01-01', '%Y-%m-%d')
            end = datetime.strptime('2022-06-01', '%Y-%m-%d')
            data = pdr.get_data_yahoo(i, start, end).reset_index()
            data['DATE'] = data['Date'].apply(lambda x: x.strftime('%Y-%m-%d'))
            data.columns = ["date","open","high","low","close",'Adj Close',"volume",'DATE']
#             data1 = data['Close'].values
        data = data.dropna(how='any',axis=0)
        forecast_out = 60
        data['prediction'] = data[['close']].shift(-forecast_out)
        X = np.array(data['close']).reshape(-1,1)
        X = X[:-forecast_out]
        y = np.array(data['prediction'])
        y = y[:-forecast_out]
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state = 7)
        model = RandomForestRegressor(n_estimators = 300, max_depth =300, random_state = 42) 
        model_fit = model.fit(X_train, y_train)
        predict = model.predict(X_test)
        score= model.score(X_test, y_test)
        realvalue=[]
        predictval=[]
        for j in range(len(X_test)):
            realvalue.append(X_test[j][0])
            predictval.append(predict[j])
        dict = {'Actual': realvalue,'Predicted':predictval}
        dfr = pd.DataFrame(dict)
        pred.append([predict])
        score = score.reshape(1,1)
        MSE=round(metrics.mean_squared_error(y_test, predict), 4)
#         RMSE=round(np.sqrt(metrics.mean_squared_error(y_test, predict)), 4)
#         R2=round(metrics.r2_score(y_test, predict), 4)
        MAPE = smape_kun(y_test, predict)
        out.append([i,RMSE,MAPE])
        data = data.join(dfr.set_index('Actual'), on='close')
        df1 = data.dropna(how='any',axis=0)
        del df1['prediction']
        df1.to_csv("C:/Users/abhi/Desktop/ARIMA_All/ARIMA_All/results_RF/"+i+"_results.csv")
#             print(data)
```
